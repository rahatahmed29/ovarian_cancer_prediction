```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost

from xgboost.callback import EarlyStopping

import seaborn as sns
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_curve, auc, precision_recall_curve, 
                           average_precision_score, accuracy_score,
                           precision_score, recall_score, f1_score,
                           roc_auc_score)
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

# Set style for plots
plt.style.use('default')
sns.set_palette("husl")

# Load the data
df = pd.read_csv('stableSet.csv')

print("Dataset shape:", df.shape)
print("\nFirst few rows:")
df.head()
print(xgboost.__version__)

# Display all column names to identify the target
print("All column names:")
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")

# Assuming the last column is the target
target_column = df.columns[-1]
print(f"\nTarget column: '{target_column}'")

# Separate features and target
X = df.drop(columns=[target_column])
y = df[target_column]

print(f"\nFeatures shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Check if target is categorical or needs conversion
print(f"\nTarget variable type: {y.dtype}")
print("Unique values in target:", y.unique())
print("Number of unique values:", len(y.unique()))

# Determine classification type
if len(y.unique()) > 2:
    print("Multi-class classification detected")
    classification_type = "multi-class"
else:
    print("Binary classification detected")
    classification_type = "binary"
    # Convert to binary if not already 0/1
    if set(y.unique()) != {0, 1}:
        y = y.astype(int)
        print("Converted target to binary integers")
# Set up 5-fold stratified cross-validation
n_splits = 5
if classification_type == "binary":
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
else:
    # For multi-class, we'll use regular KFold
    from sklearn.model_selection import KFold
    skf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store results
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
roc_auc_scores = [] if classification_type == "binary" else []
pr_auc_scores = [] if classification_type == "binary" else []
all_y_true = []
all_y_pred = []
all_y_proba = [] if classification_type == "binary" else []

# NEW: Initialize dictionaries for misclassified samples
misclassified_samples = {}
false_positives_by_fold = {}
false_negatives_by_fold = {}




# Set objective based on classification type

# -------------------------------------------------------------
#  STEP: Add Hyperparameter Tuning via GridSearchCV (before CV)
# -------------------------------------------------------------
# -------------------------------------------------------------
#  FIXED XGBOOST PARAMETERS (FROM STABILITY REPORT)
# -------------------------------------------------------------
# xgb_params = {
#     'n_estimators': int(round(377.6661)),   # ~378
#     'max_depth': int(round(5.8418)),        # ~6
#     'learning_rate': 0.0486,
#     'subsample': 0.8839,
#     'colsample_bytree': 0.8898,
#     'min_child_weight': 1.8206,
#     'gamma': 0.4182,
#     'objective': 'binary:logistic',
#     'use_label_encoder': False,
#     'eval_metric': 'auc',
#     'random_state': 42
# }
xgb_params = {
    'n_estimators': 498,   # ~378
    'max_depth': 7,        # ~6
    'learning_rate': 0.07,
    'subsample': 0.96,
    'colsample_bytree': 0.96,
    'min_child_weight': 2.62,
    'gamma': 0.76,
    'objective': 'binary:logistic',
    'use_label_encoder': False,
    'eval_metric': 'auc',
    'random_state': 42
}


xgb_params.update({
    'objective': 'binary:logistic',
    'use_label_encoder': False,
    'eval_metric': 'auc',
    'random_state': 42
})
if classification_type == "binary":
    xgb_params['objective'] = 'binary:logistic'
    xgb_params['eval_metric'] = 'auc'
else:
    xgb_params['objective'] = 'multi:softprob'
    xgb_params['eval_metric'] = 'mlogloss'
    xgb_params['num_class'] = len(y.unique())
print(f"Starting {n_splits}-fold cross-validation with misclassified sample tracking...")
# Perform 5-fold CV
for fold, (train_idx, test_idx) in enumerate(skf.split(X, y)):
    print(f"\n{'='*50}")
    print(f"Fold {fold + 1}/{n_splits}")
    print(f"{'='*50}")
    
    # Split data
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # Store original indices for reference
    test_indices = X_test.index
    
    # Create CatBoost pools
    model = XGBClassifier(**xgb_params)
    model.fit(
    X_train, y_train,
    
)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)
    
    # Store for overall evaluation
    all_y_true.extend(y_test)
    all_y_pred.extend(y_pred)
    
    if classification_type == "binary":
        y_proba_binary = y_proba[:, 1]
        all_y_proba.extend(y_proba_binary)
    
    # NEW: Identify misclassified samples
    misclassified_mask = (y_test != y_pred)
    misclassified_indices = test_indices[misclassified_mask]
    
    # NEW: Separate False Positives and False Negatives
    fp_mask = (y_test == 0) & (y_pred == 1)
    fn_mask = (y_test == 1) & (y_pred == 0)
    
    false_positive_indices = test_indices[fp_mask]
    false_negative_indices = test_indices[fn_mask]
    
    # NEW: Store misclassified samples data
    misclassified_samples[fold] = {
        'indices': misclassified_indices.tolist(),
        'X_values': X.loc[misclassified_indices],
        'y_true': y.loc[misclassified_indices],
        'y_pred': y_pred[misclassified_mask],
        'y_proba': y_proba_binary[misclassified_mask] if classification_type == "binary" else None
    }
    
    false_positives_by_fold[fold] = false_positive_indices.tolist()
    false_negatives_by_fold[fold] = false_negative_indices.tolist()
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    
    # Store scores
    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)
    
    if classification_type == "binary":
        roc_auc = roc_auc_score(y_test, y_proba_binary)
        pr_auc = average_precision_score(y_test, y_proba_binary)
        roc_auc_scores.append(roc_auc)
        pr_auc_scores.append(pr_auc)
    
    # Print fold results
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    
    # NEW: Print misclassification info
    print(f"Misclassified samples: {len(misclassified_indices)}")
    print(f"False Positives: {len(false_positive_indices)}")
    print(f"False Negatives: {len(false_negative_indices)}")
    
    if classification_type == "binary":
        print(f"ROC AUC: {roc_auc:.4f}")
        print(f"PR AUC: {pr_auc:.4f}")
    
    # Classification report
    print("\nClassification Report:")
    if classification_type == "binary":
        print(classification_report(y_test, y_pred, target_names=['Class 0', 'Class 1']))
    else:
        print(classification_report(y_test, y_pred))
    
    # Confusion matrix for this fold
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(cm)
    
    # Plot ROC curve for this fold
    if classification_type == "binary":
        plt.figure(figsize=(8, 6))
        fpr_fold, tpr_fold, _ = roc_curve(y_test, y_proba_binary)
        roc_auc_fold = auc(fpr_fold, tpr_fold)
        
        plt.plot(fpr_fold, tpr_fold, label=f'Fold {fold + 1} (AUC = {roc_auc_fold:.3f})', linewidth=2)
        plt.plot([0, 1], [0, 1], 'k--', linewidth=1, alpha=0.5)
        plt.fill_between(fpr_fold, tpr_fold, alpha=0.3)
        
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title(f'ROC Curve - Fold {fold + 1}', fontsize=14)
        plt.legend(loc='lower right', fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.xlim([0, 1])
        plt.ylim([0, 1])
        plt.tight_layout()
        plt.show()
# =============================================================================
# SHAP ANALYSIS FOR MODEL INTERPRETABILITY
# =============================================================================
print(f"\n{'='*60}")
print("SHAP ANALYSIS - MODEL INTERPRETABILITY")
print(f"{'='*60}")

try:
    import shap
    
    # Train a final model on all data for SHAP analysis
    print("Training final model for SHAP analysis...")
    final_model = XGBClassifier(**xgb_params)
    final_model.fit(X, y, verbose=False)
    
    # Create SHAP explainer
    explainer = shap.TreeExplainer(final_model)
    
    # Calculate SHAP values (use a sample for faster computation)
    sample_size = min(100, len(X))
    X_sample = X.sample(sample_size, random_state=42)
    shap_values = explainer.shap_values(X_sample)
    
    # 1. Global Feature Importance (Bar Plot)
    print("\n1. Global SHAP Feature Importance:")
    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_values, X_sample, plot_type="bar", show=False)
    plt.title("Global Feature Importance (SHAP Values)", fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()
    
    # 2. Detailed SHAP Summary Plot
    print("2. Detailed SHAP Summary:")
    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_values, X_sample, show=False)
    plt.title("SHAP Value Distribution by Feature", fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()
    
    # 3. Individual prediction explanations
    print("3. Individual Prediction Explanations (first 3 samples):")
    for i in range(min(3, sample_size)):
        plt.figure(figsize=(12, 4))
        shap.force_plot(explainer.expected_value, shap_values[i], X_sample.iloc[i], 
                       matplotlib=True, show=False)
        plt.title(f"SHAP Explanation for Sample {i+1} (True class: {y.iloc[X_sample.index[i]]})", 
                 fontsize=14, pad=20)
        plt.tight_layout()
        plt.show()
        
        # Print feature contributions
        print(f"\nSample {i+1} - Top contributing features:")
        sample_contributions = pd.DataFrame({
            'Feature': X.columns,
            'SHAP_Value': shap_values[i],
            'Abs_SHAP': np.abs(shap_values[i])
        }).sort_values('Abs_SHAP', ascending=False).head(5)
        
        print(sample_contributions.to_string(index=False))
    
except ImportError:
    print("SHAP not installed. Install with: pip install shap")
except Exception as e:
    print(f"SHAP analysis failed: {e}")
# =============================================================================
# PARTIAL DEPENDENCE PLOTS
# =============================================================================
print(f"\n{'='*60}")
print("PARTIAL DEPENDENCE ANALYSIS")
print(f"{'='*60}")

try:
    from sklearn.inspection import PartialDependenceDisplay
    
    # Get top features from your feature importance
    feature_importance = final_model.feature_importances_
    importance_df = pd.DataFrame({
        'Feature': X.columns,
        'Importance': feature_importance
    }).sort_values('Importance', ascending=False)
    
    top_features = importance_df['Feature'].head(4).tolist()
    
    print(f"Top features for partial dependence analysis: {top_features}")
    
    # Create partial dependence plots
    fig, ax = plt.subplots(2, 2, figsize=(15, 12))
    ax = ax.ravel()
    
    for i, feature in enumerate(top_features):
        PartialDependenceDisplay.from_estimator(
            final_model, X, features=[feature],
            grid_resolution=20, ax=ax[i]
        )
        ax[i].set_title(f'Partial Dependence for {feature}', fontsize=12)
        ax[i].set_ylabel('Partial Dependence')
        ax[i].grid(True, alpha=0.3)
    
    plt.suptitle("Partial Dependence Plots for Top 4 Features", fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Print insights
    print("\n Insights from Partial Dependence:")
    print(" Higher values indicate stronger positive contribution")
    print(" Lower values indicate negative contribution")
    print(" Flat lines suggest little predictive power")
    print(" Curved lines suggest complex relationships")
    
except ImportError:
    print("Partial dependence requires scikit-learn >= 0.22")
except Exception as e:
    print(f"Partial dependence analysis failed: {e}")
# NEW CELL: Functions for analyzing misclassified samples
def analyze_misclassified_samples(misclassified_samples_dict):
    """Analyze and display misclassified samples across all folds"""
    
    all_misclassified = []
    
    for fold, data in misclassified_samples_dict.items():
        print(f"\n Fold {fold + 1} Misclassified Samples:")
        print(f"   Total: {len(data['indices'])} samples")
        
        # Create DataFrame for this fold's misclassified samples
        fold_df = data['X_values'].copy()
        fold_df['Actual_Label'] = data['y_true'].values
        fold_df['Predicted_Label'] = data['y_pred']
        if data['y_proba'] is not None:
            fold_df['Prediction_Probability'] = data['y_proba']
        fold_df['Fold'] = fold + 1
        fold_df['Error_Type'] = fold_df.apply(
            lambda x: 'False Positive' if x['Actual_Label'] == 0 else 'False Negative', 
            axis=1
        )
        
        all_misclassified.append(fold_df)
        
        # Display sample details
        fp_count = len(fold_df[fold_df['Error_Type'] == 'False Positive'])
        fn_count = len(fold_df[fold_df['Error_Type'] == 'False Negative'])
        print(f"   False Positives: {fp_count}")
        print(f"   False Negatives: {fn_count}")
        print(f"   Sample indices: {data['indices'][:5]}..." if len(data['indices']) > 5 else data['indices'])
    
    # Combine all folds
    if all_misclassified:
        all_misclassified_df = pd.concat(all_misclassified)
        return all_misclassified_df
    else:
        return pd.DataFrame()

def create_error_analysis_report(misclassified_samples_dict, X, y):
    """Generate comprehensive error analysis report"""
    
    report = {
        'total_errors': 0,
        'false_positives': 0,
        'false_negatives': 0,
        'error_by_fold': {},
        'common_features_errors': {},
        'hardest_samples': []
    }
    
    all_error_indices = []
    
    for fold, data in misclassified_samples_dict.items():
        report['error_by_fold'][fold+1] = len(data['indices'])
        report['total_errors'] += len(data['indices'])
        report['false_positives'] += len([i for i in data['indices'] if y.loc[i] == 0])
        report['false_negatives'] += len([i for i in data['indices'] if y.loc[i] == 1])
        all_error_indices.extend(data['indices'])
    
    # Find most common error features
    if all_error_indices:
        error_samples = X.loc[all_error_indices]
        report['common_features_errors'] = error_samples.mean().sort_values(ascending=False).head(5).to_dict()
        
        # Find hardest samples (misclassified multiple times)
        from collections import Counter
        error_counts = Counter(all_error_indices)
        report['hardest_samples'] = [idx for idx, count in error_counts.items() if count > 1]
    
    return report
# NEW CELL: Analyze the misclassified samples
print(f"\n{'='*60}")
print("MISCLASSIFIED SAMPLE ANALYSIS")
print(f"{'='*60}")

# Analyze misclassified samples
all_misclassified_df = analyze_misclassified_samples(misclassified_samples)

if not all_misclassified_df.empty:
    # Generate error report
    error_report = create_error_analysis_report(misclassified_samples, X, y)
    
    print(f"\n Overall Error Analysis:")
    print(f"   Total misclassified samples: {error_report['total_errors']}")
    print(f"   False Positives: {error_report['false_positives']}")
    print(f"   False Negatives: {error_report['false_negatives']}")
    
    print(f"\n Errors by fold:")
    for fold, count in error_report['error_by_fold'].items():
        print(f"   Fold {fold}: {count} errors")
    
    if error_report['hardest_samples']:
        print(f"\n Hardest samples (misclassified multiple times): {error_report['hardest_samples']}")
    
    # Export to CSV
    all_misclassified_df.to_csv('misclassified_samples_analysis.csv', index=True)
    print(f"\n Misclassified samples exported to 'misclassified_samples_analysis.csv'")
else:
    print("No misclassified samples found!")
print(f"\n{'='*60}")
print("OVERALL 5-FOLD CROSS-VALIDATION RESULTS")
print(f"{'='*60}")

# Calculate mean and std for all metrics
metrics_data = {
    'Accuracy': accuracy_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1-Score': f1_scores
}

if classification_type == "binary":
    metrics_data['ROC AUC'] = roc_auc_scores
    metrics_data['PR AUC'] = pr_auc_scores

print("\nMean  Std across 5 folds:")
for metric_name, scores in metrics_data.items():
    mean_score = np.mean(scores)
    std_score = np.std(scores)
    print(f"{metric_name}: {mean_score:.4f}  {std_score:.4f}")

# Additional statistics
print(f"\nAdditional Statistics:")
print(f"Median Accuracy: {np.median(accuracy_scores):.4f}")
print(f"Range of Accuracy: {np.min(accuracy_scores):.4f} - {np.max(accuracy_scores):.4f}")
print(f"Range of F1-Score: {np.min(f1_scores):.4f} - {np.max(f1_scores):.4f}")

# Detailed statistics for all metrics
print(f"\nDetailed Statistics for All Metrics:")
print("-" * 50)
for metric_name, scores in metrics_data.items():
    mean_score = np.mean(scores)
    median_score = np.median(scores)
    min_score = np.min(scores)
    max_score = np.max(scores)
    std_score = np.std(scores)
    print(f"{metric_name}:")
    print(f"  Mean: {mean_score:.4f}")
    print(f"  Median: {median_score:.4f}")
    print(f"  Std: {std_score:.4f}")
    print(f"  Range: {min_score:.4f} - {max_score:.4f}")
    print()
# Overall classification report
print("\nOverall Classification Report (All Folds Combined):")
if classification_type == "binary":
    print(classification_report(all_y_true, all_y_pred, target_names=['Class 0', 'Class 1']))
else:
    print(classification_report(all_y_true, all_y_pred))

# Overall confusion matrix
overall_cm = confusion_matrix(all_y_true, all_y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(overall_cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=sorted(np.unique(all_y_true)),
            yticklabels=sorted(np.unique(all_y_true)))
plt.title('Confusion Matrix (XGBoost)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
if classification_type == "binary":
    # Overall ROC curve
    fpr, tpr, _ = roc_curve(all_y_true, all_y_proba)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, label=f'Overall ROC (AUC =0.9536)', linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
    plt.fill_between(fpr, tpr, alpha=0.3)

    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curve (XGBoost)', fontsize=14)
    plt.legend(loc='lower right', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.show()
if classification_type == "binary":
    # Precision-Recall Curve
    precision_curve, recall_curve, _ = precision_recall_curve(all_y_true, all_y_proba)
    pr_auc = auc(recall_curve, precision_curve)

    plt.figure(figsize=(10, 6))
    plt.plot(recall_curve, precision_curve, label=f'Overall PR (AUC = {pr_auc:.3f})', linewidth=2)
    plt.fill_between(recall_curve, precision_curve, alpha=0.3)

    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.title('Overall Precision-Recall Curve (5-Fold CV)', fontsize=14)
    plt.legend(loc='upper right', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.show()
# Train final model on all data for feature importance
final_model = XGBClassifier(**xgb_params)
final_model.fit(X, y, verbose=False)

# Feature Importance
feature_importance = final_model.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importance
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(12, 8))
bars = plt.barh(importance_df['Feature'][:15], importance_df['Importance'][:15])
plt.xlabel('Importance', fontsize=12)
plt.title('Top 15 Feature Importance', fontsize=14)
plt.gca().invert_yaxis()

# Add value labels on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.001, bar.get_y() + bar.get_height()/2, 
             f'{width:.3f}', ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()

print("Top 10 Most Important Features:")
print("=" * 40)
for i, row in importance_df.head(10).iterrows():
    print(f"{row['Feature']}: {row['Importance']:.4f}")
# Final summary
print("FINAL 5-FOLD CROSS-VALIDATION SUMMARY")
print("=" * 50)
print(f"Dataset: {X.shape[0]} samples, {X.shape[1]} features")
print(f"Target variable: '{target_column}'")
print(f"Classification type: {classification_type}")
print(f"Number of folds: {n_splits}")

print(f"\nCROSS-VALIDATION PERFORMANCE (Mean  Std):")
print("-" * 40)
for metric_name, scores in metrics_data.items():
    mean_score = np.mean(scores)
    std_score = np.std(scores)
    print(f"{metric_name}: {mean_score:.4f}  {std_score:.4f}")

print(f"\nPERFORMANCE INTERPRETATION:")
print("-" * 30)
mean_accuracy = np.mean(accuracy_scores)
if mean_accuracy > 0.9:
    print(" Excellent performance with high consistency across folds")
elif mean_accuracy > 0.8:
    print(" Very good performance")
elif mean_accuracy > 0.7:
    print(" Good performance")
elif mean_accuracy > 0.6:
    print(" Acceptable performance")
else:
    print(" Needs improvement")

if classification_type == "binary":
    mean_roc_auc = np.mean(roc_auc_scores)
    if mean_roc_auc > 0.9:
        print(" Outstanding discrimination ability")
    elif mean_roc_auc > 0.8:
        print(" Excellent discrimination ability")
    elif mean_roc_auc > 0.7:
        print(" Good discrimination ability")
```
